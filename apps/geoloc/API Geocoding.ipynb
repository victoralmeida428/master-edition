```python
# Corrigindo erros
df.loc[df['ID']==13574, 'Latitude'] = -15.7584809
df.loc[df['ID']==13574, 'Longitude'] = -47.88494795
df.head(3)
```

```python
import plotly.graph_objects as go

# Cria o mapa usando o Plotly
fig = go.Figure(go.Scattermapbox(
    lat=df['Latitude'],
    lon=df['Longitude'],
    mode='markers',
    marker=go.scattermapbox.Marker(
        size=9,
        color='rgb(0, 100, 58)',
        opacity=0.7
    ),
    text=df['Nome fantasia'],
))

# Configura o layout do mapa
fig.update_layout(
    mapbox_style='open-street-map',
    mapbox_center_lon=0,
    margin={'r': 0, 't': 0, 'l': 0, 'b': 0}
)

# Exibe o mapa
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# Lendo arquivo csv
df = pd.read_csv(r'X:\MalaDiretaCoordenadas.csv', encoding='latin1', sep=';')

for coluna in ['Latitude', 'Longitude']:
    df[coluna] = df[coluna].str.replace(',', '.')
    df[coluna] = pd.to_numeric(df[coluna])


# Cria o mapa usando o Plotly
fig = go.Figure(go.Scattermapbox(
    lat=df['Latitude'],
    lon=df['Longitude'],
    mode='markers',
    marker=go.scattermapbox.Marker(
        size=9,
        color='rgb(0, 100, 58)',
        opacity=0.7
    ),
    text=df['Nome fantasia'],
))

# Configura o layout do mapa
fig.update_layout(
    mapbox_style='open-street-map',
    mapbox_center_lon=0,
    margin={'r': 0, 't': 0, 'l': 0, 'b': 0}
)

# Exibe o mapa
fig.show()
```
